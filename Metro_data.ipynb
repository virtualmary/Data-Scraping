```python
import requests
from selenium import webdriver
import pandas as pd  
```

```python
url ='https://msk.metro-cc.ru/virtual/kachestvo?page='
```

```python
response = requests.get(url)
print(response.status_code)
```

```python
driver = webdriver.Safari()
```

```python
data = []
for n in range(1, 6):
    driver.get(url+str(n))
    result = driver.find_elements_by_class_name('catalog-item__top')
    for el in result:
        product = [el.find_element_by_class_name('catalog-item_text').find_element_by_class_name('catalog-item_name').text
                   ,float(el.find_element_by_xpath("//div[@class='catalog-item_price-old']").text[4:-2])
                   ,float(el.find_element_by_xpath("//div[@class='catalog-item_price-lvl_current']").text[4:-2])
                  ]
        data.append(product)
```

```python
data
```

```python
df = pd.DataFrame(data, columns =['name', 'old price', 'curr price']) 
df 
```

```python
df.to_csv('prices.csv')
```
